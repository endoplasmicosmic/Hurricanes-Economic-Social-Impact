```python
import pandas as pd
import geojson
```

```python
df = pd.read_csv('data.csv')
```

```python
df
```

```python
features = []
for _, row in df.iterrows():
    feature = {
        "type": "Feature",
        "geometry": {
            "type": "Point",
            "coordinates": [row["Longitude"], row["Latitude"]]
        },
        "properties": {
            "Location": row["Location"],
            "Total Deaths": row["Total Deaths"],
            "No. Affected": row["No. Affected"],
            "Total Affected": row["Total Affected"],
            "Insured Damage ('000 US$)": row["Insured Damage ('000 US$)"],
            "Total Damage ('000 US$)": row["Total Damage ('000 US$)"],
            "Magnitude": row["Magnitude"],
            "Magnitude Scale": row["Magnitude Scale"]
        }
    }
    features.append(feature)
```

```python
geojson_data = {
    "type": "FeatureCollection",
    "features": features
}
```

```python
with open('data.geojson', 'w') as f:
    geojson.dump(geojson_data, f)
```
