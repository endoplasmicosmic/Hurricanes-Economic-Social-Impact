# Analysis of Total Deaths from Hurricanes by Year

This analysis focuses on the total number of deaths caused by hurricanes each year. The data is grouped by year, and the total deaths are summed to visualize the trend over time. The resulting plot highlights the years with the highest death tolls, helping us understand the social impact of hurricanes across different time periods.


```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter
```

```python
# Load the Yearly EMDAT Data
yearly_data = pd.read_csv('Resource/yearly_emdat_data.csv')
yearly_data.head()
```

```python
# Group the data by 'Start Year' and sum 'Total Deaths' to analyze deaths by year
deaths_by_year = yearly_data.groupby('Start Year')['Total Deaths'].sum()
deaths_by_year
```

```python
# Plot the total deaths by year
plt.figure(figsize=(10, 6))
plt.plot(deaths_by_year.index, deaths_by_year.values, marker='o', color='skyblue')
plt.title('Total Deaths from Hurricanes by Year')
plt.xlabel('Year')
plt.ylabel('Total Deaths')
plt.grid(True)
plt.tight_layout()
#plt.savefig('Images/Total_Deaths_by_Year.png')
plt.show()
```

# Analysis of Most Affected Locations by Hurricanes

This analysis identifies the locations most affected by hurricanes, based on the total number of affected individuals. The data is grouped by location, and the total number of affected people is summed to highlight the areas with the greatest impact.


```python
# Load the  Data
data = pd.read_csv('Resource/emdat_data_map.csv')
data.head()
```

```python
# Replace the specific location name using loc[] to avoid SettingWithCopyWarning
data.loc[data['Location'] == 'Acapulco (Central Guerrero State)', 'Location'] = 'Acapulco (C.G.S)'
data
```

```python
# Group the data by 'Location' and sum 'Total Affected' to analyze the most affected locations
affected_by_location = data.groupby('Location')['Total Affected'].sum()
affected_by_location
```

```python
# Sort the data to find the most affected locations
affected_by_location_sorted = affected_by_location.sort_values(ascending=False).head(10)
affected_by_location_sorted
```

```python
# Plot the most affected locations
plt.figure(figsize=(15, 8))
ax = affected_by_location_sorted.plot(kind='bar', color='coral')
plt.title('Top 10 Most Affected Locations by Hurricanes') 
plt.xlabel('Location')
plt.ylabel('Total Affected') 
plt.xticks(rotation=45, ha='right')  
ax.yaxis.set_major_formatter(FuncFormatter(lambda x, pos: f'{int(x):,}'))  
plt.grid(True) 
plt.tight_layout()  
#plt.savefig('Images/Most_Affected_Locations.png')
plt.show() 
```

# Combined Analysis of Most Affected Locations by Hurricanes

This analysis identifies the top 10 locations most affected by hurricanes, based on both the total number of affected individuals and the total number of deaths. This combined view helps in understanding the overall social impact of hurricanes in different areas.


```python
map_data = pd.read_csv('Resource/emdat_data_map.csv')
map_data
```

```python
# Replace the specific location name using loc[] to avoid SettingWithCopyWarning
map_data.loc[map_data['Location'] == 'Acapulco (Central Guerrero State)', 'Location'] = 'Acapulco (C.G.S)'
map_data
```

```python
# Group the data by 'Location' and sum 'Total Affected' and 'Total Deaths'
combined_impact_by_location = map_data.groupby('Location').agg({
    'Total Affected': 'sum',
    'Total Deaths': 'sum'
})

# Sort the data by 'Total Affected' to find the top 10 most affected locations
combined_impact_sorted = combined_impact_by_location.sort_values('Total Affected', ascending=False).head(10)
combined_impact_sorted
```

```python
# Plot the most affected locations with both metrics
plt.figure(figsize=(15, 8))
ax = combined_impact_sorted[['Total Affected', 'Total Deaths']].plot(kind='bar', ax=plt.gca(), color=['coral', 'skyblue'])
plt.title('Top 10 Most Affected Locations by Hurricanes (Total Affected & Total Deaths)')
plt.xlabel('Locations')
plt.ylabel('Total Affected & Total Deaths')
plt.xticks(rotation=45, ha='right')
ax.yaxis.set_major_formatter(FuncFormatter(lambda x, pos: f'{int(x):,}'))  
plt.grid(True) 
plt.tight_layout()
plt.savefig('Images/Most_Affected_Locations_Combined.png')
plt.show()
```
