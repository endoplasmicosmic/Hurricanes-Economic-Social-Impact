```python
import json
import geopandas as gpd
from shapely.geometry import Point
```

```python
emdat_data_map = pd.read_csv('../Resource/emdat_data_map.csv')
emdat_data_map
```

```python
# Ensure that 'Longitude' and 'Latitude' columns are numeric
emdat_data_map['Longitude'] = pd.to_numeric(emdat_data_map['Longitude'], errors='coerce')
emdat_data_map['Latitude'] = pd.to_numeric(emdat_data_map['Latitude'], errors='coerce')

# Drop rows with missing values in 'Longitude' or 'Latitude'
emdat_data_map = emdat_data_map.dropna(subset=['Longitude', 'Latitude'])

# Create GeoDataFrame
geometry = [Point(xy) for xy in zip(emdat_data_map['Longitude'], emdat_data_map['Latitude'])]
geo_df = gpd.GeoDataFrame(emdat_data_map, geometry=geometry)

# Save to GeoJSON
geo_df.to_file('hurricane_data.geojson', driver='GeoJSON')
```
